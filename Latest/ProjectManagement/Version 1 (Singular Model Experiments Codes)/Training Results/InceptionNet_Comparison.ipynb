```python
import pandas as pd
import matplotlib.pyplot as plt

Inception40X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNetResized40X_CalledOnce.csv'

Inception100X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNet100X.csv'

Inception200X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNet200X.csv'

Inception400X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNet400X.csv'

Inception40XResized = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNetResized40X_CalledTwice.csv'

Inception100XResized = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNetResized100X_CalledTwice.csv'

Inception200XResized = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNetResized200X_CalledTwice.csv'

Inception400XResized = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Level6\Final Year Project\Z. ProgressBackup\Model Training Results\CSV_History\InceptionNetResized400X_CalledTwice.csv'




# 
data_40X = pd.read_csv(Inception40X, nrows=100)
data_100X = pd.read_csv(Inception100X)
data_200X = pd.read_csv(Inception200X)
data_400X = pd.read_csv(Inception400X)
data_40X_resized = pd.read_csv(Inception40XResized)
data_100X_resized = pd.read_csv(Inception100XResized)
data_200X_resized = pd.read_csv(Inception200XResized)
data_400X_resized = pd.read_csv(Inception400XResized)

# #  ResNet 
# resnet_data.rename(columns=lambda x: x.replace('_1', ''), inplace=True)

# 
def plot_metric_comparison_all_models(metric):
    plt.figure(figsize=(12, 6))

    # 
    plt.plot(data_40X['epoch'][:100], data_40X[metric][:100], label='Inception 40X', marker='x')
    plt.plot(data_100X['epoch'], data_100X[metric], label='Inception 100X', marker='o')
    plt.plot(data_200X['epoch'], data_200X[metric], label='Inception 200X', marker='^')
    plt.plot(data_400X['epoch'], data_400X[metric], label='Inception 400X', marker='s')
    plt.plot(data_40X_resized['epoch'], data_40X_resized[metric], label='Inception 40X Resized', marker='x')
    plt.plot(data_100X_resized['epoch'], data_100X_resized[metric], label='Inception 100X Resized', marker='+')
    plt.plot(data_200X_resized['epoch'], data_200X_resized[metric], label='Inception 200X Resized', marker='*')
    plt.plot(data_400X_resized['epoch'], data_400X_resized[metric], label='Inception 400X Resized', marker='d')


    plt.title(f'Comparison of {metric} Across Models')
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.show()

# 
metrics = ['accuracy', 'loss','f1_score', 'precision', 'recall', 'auc',
          'val_accuracy', 'val_loss','val_f1_score', 'val_precision', 'val_recall','val_auc']
for metric in metrics:
    plot_metric_comparison_all_models(metric)
```
