```python
import pandas as pd
import matplotlib.pyplot as plt

Inception100X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\大四下\Final Year Project\Z. ProgressBackup\CSV_History\Comparison\InceptionV3.csv'

Inception200X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\大四下\Final Year Project\Z. ProgressBackup\CSV_History\Comparison\ResNet50V2.csv'

Inception400X = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\大四下\Final Year Project\Z. ProgressBackup\CSV_History\FinalModifyCSV\Version8_run5-DIR(Best at 2024 3 13).csv'
```

```python

# 加载数据
data_100X = pd.read_csv(Inception100X)
data_200X = pd.read_csv(Inception200X)
data_400X = pd.read_csv(Inception400X)
```

```python
def plot_metric_comparison_all_models(metric):
    plt.figure(figsize=(12, 6))

    # 为每个模型绘制图表
    plt.plot(data_100X['epoch'], data_100X[metric], label='InceptionV3', marker='o')
    plt.plot(data_200X['epoch'], data_200X[metric], label='ResNet50V2', marker='^')
    plt.plot(data_400X['epoch'], data_400X[metric], label='DIRA-Net', marker='s')
    
    
    

    plt.title(f'Comparison of {metric} Across Models')
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

# 对每个指标绘制比较图
metrics = [
          'val_accuracy', 'val_loss','val_f1_score', 'val_precision', 'val_recall','val_auc']
for metric in metrics:
    plot_metric_comparison_all_models(metric)
```
