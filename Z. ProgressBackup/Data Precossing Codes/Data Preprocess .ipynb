```python
import os
import shutil

# 设置文件夹路径
folder_path = '/root/Breast Cancer/Datasets/40X/train/malignant'  # 请替换为你的图片文件夹路径

target_num_images = 1300  # 目标图片数量

while True:
    files = [file for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]
    current_num_images = len(files)
    if current_num_images >= target_num_images:
        break  # 如果达到或超过目标数量，则停止

    # 计算还需要复制的图片数量
    num_images_to_copy = min(target_num_images - current_num_images, current_num_images)

    for file in files[:num_images_to_copy]:
        original_path = os.path.join(folder_path, file)
        new_path = os.path.join(folder_path, f"aug_{current_num_images}_{file}")
        shutil.copy2(original_path, new_path)
        current_num_images += 1
        if current_num_images >= target_num_images:
            break
```
